```python
df["Selling Price"] = (df["Netflix Global Users"] / df["Global Revenue"]) * 100
df
```

```python
df.loc[(df["Date"].dt.month >= 3) & (df["Date"].dt.month <= 5), "Season"] = "Spring"
df.loc[(df["Date"].dt.month >= 6) & (df["Date"].dt.month <= 8), "Season"] = "Summer"
df.loc[(df["Date"].dt.month >= 9) & (df["Date"].dt.month <= 11), "Season"] = "Autumn"
df.loc[(df["Date"].dt.month >= 12) | (df["Date"].dt.month <= 2), "Season"] = "Winter"
```

```python
df.columns
```

```python
df.groupby(df["Season"]).mean()["Selling Price"].plot()
```

```python
df.groupby(df["Season"]).mean()["Netflix Global Users"].plot()
```

```python
df["Demos"] = df["Domestic Free Trialers"] + df["Interntaional Free Trialers"]
```
